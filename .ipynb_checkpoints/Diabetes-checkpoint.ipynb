```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x = data.drop('Outcome', axis = 1) # x here represents our features (independen variable)
y = data['Outcome'] # y here represents our label (dependent variable)
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.20, random_state=30)
```

```python
x_train
```

```python
lr = LogisticRegression() 
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
y_train.shape 
```

```python
y_test
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier(max_depth=5, min_samples_split=10, min_samples_leaf=5, criterion='gini')
```

```python
dtc.fit(x_train,y_train)
```

```python
y_pred2 = dtc.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred2)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred2))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred2))
```
